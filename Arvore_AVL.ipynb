```python
class node:
    def __init__(self, num):
        self.value = num
        self.esq = None
        self.dir = None
        self.altura = 1

class AVL:

    def altura(self, Node):
        if Node is None:
            return 0
        else:
            return Node.altura

    def balancear(self, Node):
        if Node is None:
            return 0
        else:
            return self.altura(Node.esq) - self.altura(Node.dir)

    def ValorMinNo(self, Node):
        if Node is None or Node.esq is None:
            return Node
        else:
            return self.ValorMinNo(Node.esq)

    def rotacionarDir(self, Node):
        a = Node.esq
        b = a.dir
        a.dir = Node
        Node.esq = b
        Node.altura = 1 + max(self.altura(Node.esq), self.altura(Node.dir))
        a.altura = 1 + max(self.altura(a.esq), self.altura(a.dir))
        return a

    def rotacionarEsq(self, Node):
        a = Node.dir
        b = a.esq
        a.esq = Node
        Node.dir = b
        Node.altura = 1 + max(self.altura(Node.esq), self.altura(Node.dir))
        a.altura = 1 + max(self.altura(a.esq), self.altura(a.dir))
        return a

    def inserir(self, val, raiz):
        if raiz is None:
            return node(val)
        elif val <= raiz.value:
            raiz.esq = self.inserir(val, raiz.esq)
        elif val > raiz.value:
            raiz.dir = self.inserir(val, raiz.dir)
        raiz.altura = 1 + max(self.altura(raiz.esq), self.altura(raiz.dir))
        balancear = self.balancear(raiz)
        if balancear > 1 and raiz.esq.value > val:
            return self.rotacionarDir(raiz)
        if balancear < -1 and val > raiz.dir.value:
            return self.rotacionarEsq(raiz)
        if balancear > 1 and val > raiz.esq.value:
            raiz.esq = self.rotacionarEsq(raiz.esq)
            return self.rotacionarDir(raiz)
        if balancear < -1 and val < raiz.dir.value:
            raiz.dir = self.rotacionarDir(raiz.dir)
            return self.rotacionarEsq(raiz)
        return raiz

    def preorder(self, raiz):
        if raiz is None:
          return
        print(raiz.value)
        self.preorder(raiz.esq)
        self.preorder(raiz.dir)

    def inordem(self, raiz):
        if raiz is None:
          return
        self.inordem(raiz.esq)
        print(raiz.value)
        self.inordem(raiz.dir)
        
    def posordem(self, raiz):
        if raiz is None:
          return
        self.posordem(raiz.esq)
        self.posordem(raiz.dir)
        print(raiz.value)

    def delete(self, val, Node):
        if Node is None:
            return Node
        elif val < Node.value:
            Node.esq = self.delete(val, Node.esq)
        elif val > Node.value:
            Node.dir = self.delete(val, Node.dir)
        else:
            if Node.esq is None:
                lt = Node.dir
                Node = None
                return lt
            elif Node.dir is None:
                lt = Node.esq
                Node = None
                return lt
            rgt = self.ValorMinNo(Node.dir)
            Node.value = rgt.value
            Node.dir = self.delete(rgt.value, Node.dir)
        if Node is None:
            return Node
        Node.altura = 1 + max(self.altura(Node.esq), self.altura(Node.dir))
        balancear = self.balancear(Node)
        if balancear > 1 and self.balancear(Node.esq) >= 0:
            return self.rotacionarDir(Node)
        if balancear < -1 and self.balancear(Node.dir) <= 0:
            return self.rotacionarEsq(Node)
        if balancear > 1 and self.balancear(Node.esq) < 0:
            Node.esq = self.rotacionarEsq(Node.esq)
            return self.rotacionarDir(Node)
        if balancear < -1 and self.balancear(Node.dir) > 0:
            Node.dir = self.rotacionarDir(Node.dir)
            return self.rotacionarEsq(Node)
        return Node

Tree = AVL()
rt = None
rt = Tree.inserir(1, rt)
rt = Tree.inserir(2, rt)
rt = Tree.inserir(3, rt)
rt = Tree.inserir(4, rt)
rt = Tree.inserir(5, rt)
rt = Tree.inserir(6, rt)
rt = Tree.inserir(7, rt)
rt = Tree.inserir(8, rt)
rt = Tree.inserir(9, rt)

print("PREORDER")
Tree.preorder(rt)
print("\n\nINORDEM")
Tree.inordem(rt)
print("\n\nPOSORDEM")
Tree.posordem(rt)
```
